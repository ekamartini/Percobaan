## Contoh Merubah Data Numerical ke Categorical dengan Dictionary

# Merubah data Categorical ke Numerical

```python
data_bunga['jadi numerical'] = data_bunga ['data angka'].map({'setosa':0, 'versicolor' : 1, 'virginica' : 2})
```

# Plot Hubungan Data Kolom dengan target

```python
sns.relplot(x = 'sepal length (cm)', y = 'kode bunga', hue = 'nama_bunga', data=data_bunga)
plt.show()
#hue : pembedanya
```

```python
sns.relplot(x = 'sepal width (cm)', y = 'kode bunga', hue = 'nama_bunga', data=data_bunga)
plt.show()
```

```python
sns.relplot(x = 'sepal length (cm)', y = 'kode bunga', hue = 'nama_bunga', data=data_bunga)
plt.title ('petal length (cm)', y=1,x=1)
plt.show()
```
